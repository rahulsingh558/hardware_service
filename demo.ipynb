```python
from matchbox2 import MatchBox2Laser

available_lasers = MatchBox2Laser().get_available_lasers()
laser = MatchBox2Laser()
laser.connect(available_lasers[0].portName)
```

```python
laser.set_laser_on()
readings = laser.get_laser_readings()
print(readings)
```

```python
laser.set_laser_off()
readings = laser.get_laser_readings()
print(readings)
```

```python
laser.disconnect()
```
